# Problem Statement

A bank issues credit card to its customers under the scheme of Master card. Based on past data, it is known that 60% of the customers pay the bill on time. 
If a sample of 7 accounts is selected at random from current database, construct the binomial probability distribution of accounts paying on time.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
n=7
```

```python
p=.6
```

```python
k=np.arange(0,8)
k
```

```python
binomial=stats.binom.pmf(k,n,p)
```

```python
binomial
```

```python
plt.plot(k,binomial,'o-')
plt.xlabel("Number of success")
plt.ylabel("Number of success")
plt.show()
```
